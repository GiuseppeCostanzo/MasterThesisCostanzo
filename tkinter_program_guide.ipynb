```python
import tkinter as tk

def on_button_click(name):
    print(f"Pulsante premuto per l'elemento: {name}")

root = tk.Tk()
root.title("Lista di elementi")
root.geometry("500x500")

# Lista di elementi
elementi = ["Elemento 1", "Elemento 2", "Elemento 3", "Elemento 4", "Elemento 5", "Elemento 6", "Elemento 7", "Elemento 8", "Elemento 9", "Elemento 10",
            "Elemento 11", "Elemento 12", "Elemento 13", "Elemento 14", "Elemento 15", "Elemento 16", "Elemento 17", "Elemento 18", "Elemento 19", "Elemento 20"]

# Creazione del frame per visualizzare la lista
frame = tk.Frame(root)
frame.pack(fill=tk.BOTH, expand=True)

# Visualizza ogni elemento nella lista
for row, elemento in enumerate(elementi):
    # Etichetta per il nome dell'elemento
    label = tk.Label(frame, text=elemento, width=30, anchor="w", pady=5)
    label.grid(row=row, column=0, sticky="w")

    # Pulsante associato all'elemento
    button = tk.Button(frame, text="Pulsante", command=lambda e=elemento: on_button_click(e), width=10)
    button.grid(row=row, column=1, sticky="e")

root.mainloop()
```
